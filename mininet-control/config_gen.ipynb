```python
import json
```

```python
city_abbr = {}
city_radius = {}
city_population = {}
network_stats = {}

with open('../utility/anchors-by-city/city_abbr.json') as f_city_abbr:
    city_abbr = json.load(f_city_abbr)

with open('../utility/anchors-by-city/city_radius.json') as f_city_radius:
    city_radius = json.load(f_city_radius)

with open('../utility/anchors-by-city/city_population.json') as f_city_population:
    city_population = json.load(f_city_population)

with open('../utility/anchors-by-city/network_stats.json') as f_network_stats:
    network_stats = json.load(f_network_stats)
```

```python
combined_data = {}

i = 12
for city, abbr in city_abbr.items():
    combined_data[abbr] = {
        "city": city,
        "radius": city_radius[city],
        "population": city_population[city],
        "network_stats": {},
        "ip_prefix": i
    }
    i += 4

for city_to_city, stat in network_stats.items():
    city1, city2 = city_to_city.split('|')
    city1_abbr = city_abbr[city1]
    city2_abbr = city_abbr[city2]
    combined_data[city1_abbr]["network_stats"][city2_abbr] = stat
    combined_data[city2_abbr]["network_stats"][city1_abbr] = stat
```

```python
with open('./city_config.json', 'w') as f_city_config:
    json.dump(combined_data, f_city_config, indent=4)
```

```python
import json
import math
```

```python
cities = ['Tokyo', 'Jakarta', 'Delhi', 'Mumbai', 'Manila', 'São Paulo', 'Seoul', 'Mexico City', 'New York City', 'Bangkok', 'Moscow', 'Buenos Aires', 'Lagos', 'Los Angeles', 'Ho Chi Minh City', 'Osaka', 'Johannesburg', 'Istanbul', 'Tehran']
```

```python

city_radius = {}

with open('../../info/30-largest-cities-by-population-2026.json', 'r') as file_cities:
    data = json.load(file_cities)
    for city in data:
        if city['city'] not in cities:
            continue

        area_km2 = city["area"]
        radius_km = math.sqrt(area_km2 / math.pi)
        city_radius[city['city']] = radius_km
```

```python
with open('./city_radius.json', 'w') as f_city_radius:
    json.dump(city_radius, f_city_radius, indent=4)
```

```python
with open('./city_radius.json', 'r') as f_city_radius:
    city_radius = json.load(f_city_radius)
```
